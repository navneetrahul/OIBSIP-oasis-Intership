### Remove the Outliers

- *1.iqr=q3-q1*
- *2.Value>q3+1.5*iqr*
- *3.Value<q1-1.5*iqr*

```python
df.shape
```

Now the dataset has 301 rows and 9 columns

```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)
iqr=q3-q1

upper_limit=q3+1.5*iqr
lower_limit=q1-1.5*iqr
```

```python
df1=df[~((df<(lower_limit))|(df>(upper_limit))).any(axis=1)]
df1.head()
```

```python
df1.shape
```

**After outlier treatment,the dataset has retained 264 rows and 9 columns**

```python
df1.std()
```

- As Owner has 0 std which means there is no contribution from this variable.Good to remove this variable

```python
df2=df1.drop('Owner',axis=1)
```

```python
df2.shape
```

### Encoding the categorical variables

```python
df2_cat=df2.select_dtypes(exclude=np.number)
df2_cat
```

```python
df2_cat_en=pd.get_dummies(data=df2_cat,drop_first=True)  #One hot encoding
df2_cat_en.head()
```

```python
df2_cat_en.shape
```

```python
df2_num=df2.select_dtypes(include=np.number).drop('Selling_Price',axis=1)
df2_num.head()
```

```python
df2_num.shape
```

### Concatenate the numerical and encoded categorical variable

```python
X_unsc=pd.concat([df2_num,df2_cat_en],axis=1)
X_unsc = X_unsc.reset_index(drop=True)
X_unsc.head()
```

```python
X_unsc.shape
```

### Scale the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_columns = df2_num.columns
sc = StandardScaler()

X_sc = X_unsc.copy()


X_sc[numerical_columns] = sc.fit_transform(X_unsc[numerical_columns])

X_sc.head()
```

```python
X_sc.shape
```

### Transformation

```python
y=np.log(df2['Selling_Price'])
```

```python
y = y.reset_index(drop=True)
```

```python
y.shape
```

### Modelling

```python
from sklearn.model_selection import train_test_split
```

```python
X=X_sc
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.20,random_state=1)
```

```python
xtrain.head()
```

```python
print(xtrain.shape)
print(xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

### Build the Basic Model

```python
lr=LinearRegression()

model_lr=lr.fit(xtrain,ytrain)
pred_train=model_lr.predict(xtrain)
pred_test=model_lr.predict(xtest)
```

### Alternative One - Calculating R2

```python
r2_train=r2_score(ytrain,pred_train)
r2_test=r2_score(ytest,pred_test)

print('R2 Train: ',r2_train)
print('R2 Test: ',r2_test)
```

### Alternative Two - Calculating R2

```python
print('R2 Train: ',model_lr.score(xtrain,ytrain))
print('R2 Test: ',model_lr.score(xtest,ytest))
```

R2 score of 0.946 on the training set indicates that the model is able to explain a large portion of the variability in Selling Price based on the features.
R2 score of 0.848 on the test set suggests that the model generalizes well to new, unseen data, but there is a small drop in predictive performance compared to the training set.

### Calculating RMSE Values

```python
print("RMSE Train: ",np.sqrt(mean_squared_error(ytrain,pred_train)))

print("RMSE Test: ",np.sqrt(mean_squared_error(ytest,pred_test)))
```

Lower RMSE values indicate better performance.
The RMSE values indicate that, on average, the predicted Selling Prices are close to the actual values, with slightly higher error observed in the test set compared to the training set

### Lets Test this with a Different Model like Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()

rf=rf.fit(xtrain,ytrain)

pred_train_rf=rf.predict(xtrain)

pred_test_rf=rf.predict(xtest)
```

```python
#R2 Score
r2_train=r2_score(ytrain,pred_train_rf)
r2_test=r2_score(ytest,pred_test_rf)

print('R2 Train: ',r2_train)
print('R2 Test: ',r2_test)
```

R2 scores suggest that our Random Forest model is providing a very strong fit to both the training and test data.
An R2 score of 0.9968 on the training set indicates an almost perfect fit to the training data, and the model is able to explain a very large proportion of the variability in Selling Price based on the features.
An R 2 score of 0.976 on the test set suggests that the model generalizes exceptionally well to new, unseen data.

```python
#Root Mean Squared Error
print("RMSE Train: ",np.sqrt(mean_squared_error(ytrain,pred_train_rf)))

print("RMSE Test: ",np.sqrt(mean_squared_error(ytest,pred_test_rf)))
```

A lower RMSE indicates better predictive performance. In this case, the RMSE values are quite low, suggesting that the Random Forest model is making accurate predictions

The RMSE values are relatively small, indicating that our Gradient Boosting model is making accurate predictions.
The RMSE Test value (0.161) is comparable to or slightly lower than the RMSE Test values obtained from the Random Forest Regressor we previously trained.
